```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import time

# First function using NumPy
def numpy_data_processing_function():
    data = np.random.normal(size=10000000)
    count = len(data)
    average = np.mean(data)
    total_sum = np.sum(data)
    return count, average, total_sum

# Second function using Pandas
def pandas_data_processing_function():
    data = pd.Series(np.random.normal(size=10000000))
    count = data.count()
    average = data.mean()
    total_sum = data.sum()
    return count, average, total_sum

# Measuring and storing the runtime for each function
numpy_runtimes = []
pandas_runtimes = []

for _ in range(300):
    start_time = time.time()
    numpy_data_processing_function()
    end_time = time.time()
    numpy_runtimes.append(end_time - start_time)

    start_time = time.time()
    pandas_data_processing_function()
    end_time = time.time()
    pandas_runtimes.append(end_time - start_time)

# Creating a DataFrame to store the runtimes
df_runtimes = pd.DataFrame({'NumPy Runtime': numpy_runtimes, 'Pandas Runtime': pandas_runtimes})

# Plotting the runtimes using seaborn
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
sns.histplot(df_runtimes['NumPy Runtime'], kde=True, color='blue')
plt.title('NumPy Function Runtime')
plt.xlabel('Runtime (seconds)')
plt.ylabel('Frequency')

plt.subplot(1, 2, 2)
sns.histplot(df_runtimes['Pandas Runtime'], kde=True, color='green')
plt.title('Pandas Function Runtime')
plt.xlabel('Runtime (seconds)')
plt.ylabel('Frequency')

plt.tight_layout()

# Export Plot
plt.savefig('./images/plot.png', dpi=1000, bbox_inches='tight')
```
